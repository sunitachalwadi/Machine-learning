```python
missing_values = ["n/a", "na", " "]
df = pd.read_csv(r'C:\Users\anirudh.ghosh\Desktop\telecom_churn.csv',na_values=missing_values) 
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df["TotalCharges"].fillna(df["TotalCharges"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df_num = df.select_dtypes(["int64","float64"])
df_cat = df.select_dtypes("object")
```

```python
df_num.head()
```

```python
df_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Looping each column name for encoding purpose
for col in df_cat:
    le = LabelEncoder()
    # fit_transform function converts all the words into numbers
    df_cat[col] = le.fit_transform(df_cat[col]) 
```

```python
df_cat.head()
```

```python
df_new = pd.concat([df_num,df_cat],axis=1)
```

```python
df_new.head()
```

```python
x = df_new.drop("Churn", axis=1)
y = df_new["Churn"]
```

```python
from sklearn.preprocessing import StandardScaler 
for col in x.columns:
    sc = StandardScaler()
    x[col] = sc.fit_transform(x[[col]])
    
x.head()
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
log.score(x_test,y_test)
```

```python
log.coef_
```

```python
y_pred = log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.decomposition import PCA 
```

```python
pca=PCA(n_components=4)
x_pca =pca.fit_transform(x)
print(pca.explained_variance_ratio_)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size=0.3) 
log=LogisticRegression()
log.fit(x_train,y_train)
y_pred=log.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
for i in range(18,0,-1):
    pca = PCA(n_components =i)
    x_pca=pca.fit_transform(x)
    x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size=0.3)
    log=LogisticRegression()
    log.fit(x_train,y_train)
    y_pred=log.predict(x_test)
    print("Components ={} Accuracy ={}".format(i,accuracy_score(y_test,y_pred)))
```
